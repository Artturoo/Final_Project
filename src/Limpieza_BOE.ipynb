```python
import regex as re
```

```python
def limpiar_date(fecha_de_inicio): 
    fecha_de_inicio = re.findall('^\d{2}\-\d+\-\d+', fecha_de_inicio)
    fecha_de_inicio = ''.join(fecha_de_inicio)
    return fecha_de_inicio
```
